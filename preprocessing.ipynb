```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Folder input dan output
input_folder = "dataset"
output_folder = "dataset_processing"

# Ukuran resize
resize_size = (256, 256)

# Buat folder output jika belum ada
os.makedirs(output_folder, exist_ok=True)

# Loop setiap folder kelas
for class_name in os.listdir(input_folder):
    class_input_path = os.path.join(input_folder, class_name)

    if os.path.isdir(class_input_path):

        class_output_path = os.path.join(output_folder, class_name)
        os.makedirs(class_output_path, exist_ok=True)

        # Loop semua gambar
        for filename in os.listdir(class_input_path):

            image_path = os.path.join(class_input_path, filename)

            img = cv2.imread(image_path)

            if img is not None:

                # Resize
                img_resize = cv2.resize(img, resize_size)

                # Grayscale
                img_gray = cv2.cvtColor(img_resize, cv2.COLOR_BGR2GRAY)

                # Simpan hasil
                save_path = os.path.join(class_output_path, filename)
                cv2.imwrite(save_path, img_gray)
                
print("Preprocessing selesai!")
```

```python
folder_asli = "dataset/daun_sehat_100"

files = os.listdir(folder_asli)[:4]

plt.figure(figsize=(12, 6))

for i, file in enumerate(files):

    # Gambar asli
    img_asli = cv2.imread(os.path.join(folder_asli, file))
    img_asli_rgb = cv2.cvtColor(img_asli, cv2.COLOR_BGR2RGB)

    plt.subplot(2, 4, i + 1)
    plt.imshow(img_asli_rgb)
    plt.title("Asli")
    plt.axis("off")

    # Grayscale (langsung dari gambar asli)
    img_resize = cv2.resize(img_asli, (256, 256))
    img_gray = cv2.cvtColor(img_resize, cv2.COLOR_BGR2GRAY)

    plt.subplot(2, 4, i + 5)
    plt.imshow(img_gray, cmap='gray')
    plt.title("Grayscale")
    plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
folder_asli = "dataset/Grape_black_rot_100"
folder_gray = "dataset_processing/Grape_black_rot_100"

files = sorted(os.listdir(folder_asli))[:4]

plt.figure(figsize=(12, 6))

for i, file in enumerate(files):

    # Gambar asli
    img_asli = cv2.imread(os.path.join(folder_asli, file))
    img_asli = cv2.cvtColor(img_asli, cv2.COLOR_BGR2RGB)

    plt.subplot(2, 4, i + 1)
    plt.imshow(img_asli)
    plt.title("Black Rot (Asli)")
    plt.axis("off")

    # Gambar grayscale
    img_gray = cv2.imread(
        os.path.join(folder_gray, file),
        cv2.IMREAD_GRAYSCALE
    )

    plt.subplot(2, 4, i + 5)
    plt.imshow(img_gray, cmap='gray')
    plt.title("Black Rot (Prepro)")
    plt.axis("off")

plt.tight_layout()
plt.show()
```
